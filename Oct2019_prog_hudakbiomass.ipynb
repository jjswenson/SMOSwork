```python
hud2013 = pd.read_csv(r"SMOS_hudak2013.csv")
hud2014 = pd.read_csv(r"SMOS_hudak2014.csv")
hud2016 = pd.read_csv(r"SMOS_hudak2016.csv")
lvod_explan = pd.read_csv(r"D:/Research/Texas_drought/RadarTexas/Variables/FromPython/ExplanLVODAnn10_17.csv")
```

```python
HudBiomassExplan = lvod_explan.merge(hud2013,on='ID').merge(hud2014,on='ID').merge(hud2016,on='ID')
list(HudBiomassExplan)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.relplot(x="2013MEAN", y="n2013", hue="ForPercPix414243",
           alpha=.5, height=6, data=HudBiomassExplan)
sns.relplot(x="2014MEAN", y="n2014", hue="F_evPerc42",
           alpha=.5, height=6, data=HudBiomassExplan)
plt.legend(loc='upper left')
sns.relplot(x="2013MEAN", y="NCBDMean_TperHa", alpha=.5, height=6, data=HudBiomassExplan)
#sns.regplot(x="n2014", y="PNWBiom_ForMEAN", order = 2, data=PNW_biomass_explan, ci=68)
```
